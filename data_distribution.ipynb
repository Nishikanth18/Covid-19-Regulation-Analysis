```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_excel("regulations.xlsx")
```

```python
df.info()
```

```python
fig = sns.distplot(df[' Infections']);
print(fig)
fig.savefig('file.png')
```

```python
print("Skewness: %f" % df[' Infections'].skew())
print("Kurtosis: %f" % df[' Infections'].kurt())
```

```python
#histogram and normal probability plot
sns.distplot(df[' Infections'], fit=norm);
fig = plt.figure()
res = stats.probplot(df[' Infections'], plot=plt)
```

```python
df[' Infections'] = np.log(df[' Infections'])
```

```python
#transformed histogram and normal probability plot
sns.distplot(df[' Infections'], fit=norm);
fig = plt.figure()
res = stats.probplot(df[' Infections'], plot=plt)
```

```python
print("Skewness: %f" % df[' Infections'].skew())
print("Kurtosis: %f" % df[' Infections'].kurt())
```

```python
print('Without transformation:  Skewness: 0.722639 , Kurtosis: 0.055111')
print('log     transformation:  Skewness: -0.353382, Kurtosis:-0.396644')
```
